# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
array_1d = np.zeros(10)
array_1d[4] = 1

array_2d = array_1d.reshape(2, 5)
print("Одномерный массив с пятым элементом 1, преобразованный в двумерный:")
print(array_2d)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
array_range = np.arange(10, 50)

array_reversed = array_range[::-1]

even_elements = array_reversed[array_reversed % 2 == 0]
print("\nРазвернутый массив от 10 до 49 и четные элементы:")
print(array_reversed)
print(even_elements)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
array_3x3 = np.arange(9).reshape(3, 3)
print("\nДвумерный массив 3x3:")
print(array_3x3)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
random_array = np.random.rand(4, 3, 2)

min_value = np.min(random_array)
max_value = np.max(random_array)
print(f"\nМинимум: {min_value}, Максимум: {max_value}")
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение.

```python
array_6x4 = np.random.rand(6, 4)
array_4x3 = np.random.rand(4, 3)

result_matrix = np.dot(array_6x4, array_4x3)
print("\nРезультат матричного умножения:")
print(result_matrix)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
random_7x7 = np.random.rand(7, 7)

mean_value = np.mean(random_7x7)
std_dev = np.std(random_7x7)
print(f"\nСреднее: {mean_value}, Стандартное отклонение: {std_dev}")

normalized_array = (random_7x7 - mean_value) / std_dev
print("\nНормализованный массив:")
print(normalized_array)
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], bins=30, kde=True, color='skyblue')
plt.title("Распределение общего счета")
plt.xlabel("Общий счет")
plt.ylabel("Частота")
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.title("Зависимость чаевых от общего счета")
plt.xlabel("Общий счет")
plt.ylabel("Чаевые")
plt.show()
```

### Постройте pairplot

```python
plt.figure(figsize=(12, 8))
sns.pairplot(tips, hue="sex", diag_kind="kde")
plt.suptitle("Парные графики для набора данных tips", y=1.02)
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("Коробчатая диаграмма общего счета по дням")
plt.xlabel("День")
plt.ylabel("Общий счет")
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
plt.figure(figsize=(12, 6))
for i, time in enumerate(tips['time'].unique()):
    plt.subplot(1, 2, i+1)
    sns.histplot(tips[tips['time'] == time]['tip'], bins=10, kde=True)
    plt.title(f"Распределение чаевых для {time}")
    plt.xlabel("Чаевые")
    plt.ylabel("Частота")
plt.tight_layout()
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
plt.figure(figsize=(14, 6))
for i, sex in enumerate(tips['sex'].unique()):
    plt.subplot(1, 2, i+1)
    sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips[tips['sex'] == sex])
    plt.title(f"Зависимость чаевых от общего счета для {sex}")
    plt.xlabel("Общий счет")
    plt.ylabel("Чаевые")
plt.tight_layout()
plt.show()
```
